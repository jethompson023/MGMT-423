```python
from google.colab import auth
auth.authenticate_user()

import gspread
from google.auth import default
creds, _ = default()

gc = gspread.authorize(creds)

worksheet = gc.open('ToyotaCorolla').sheet1

#get_all_values gives a list of rows
rows = worksheet.get_all_values()
```

```python
#Convert to a DataFrame and render
import pandas as pd
df = pd.DataFrame.from_records(rows)
df.columns = df.iloc[0]
df = df[1:]
df.head()
```

```python
from sklearn.model_selection import train_test_split

train_data, temp = train_test_split(df, test_size=0.5, random_state=1)
validation_data, test_data = train_test_split(temp, test_size=0.4, random_state=1)
```

```python
# Check the shape of the datasets
print(f'Expected data shapes (may be slightly off due to rounding error):')
print(f'Training data: ({int(len(df) * 0.5)}, 365)')
print(f'Validation data: ({int(len(df) * 0.3)}, 365)')
print(f'Test data: ({int(len(df) * 0.2)}, 365)')

print('\nActual data shapes:')
print(f'Training data shape: {train_data.shape}')
print(f'Validation data shape: {validation_data.shape}')
print(f'Test data shape: {test_data.shape}')
```

```python
#convert the values in the binary outputs
bin_df = pd.get_dummies(df,columns=['Fuel_Type', 'Color'])
bin_df.head()
```

```python
x = bin_df.drop('Price', axis=1)
# x = bin_df['Fuel_Type']
y = bin_df['Price']

#partition the dataset into x & y variables for training, validation, & test
x_trainval, x_test = train_test_split(x, test_size=0.2, random_state=42)
x_train, x_val = train_test_split(x_trainval, test_size=0.3, random_state=42)

y_trainval, y_test = train_test_split(y, test_size=0.2, random_state=42)
y_train, y_val = train_test_split(y_trainval, test_size=0.3, random_state=42)
```

```python
#show results for training dataset
print("Training Data Results -> n/")
print(x_train, y_train)
```

```python
#show results for training dataset
print("Validation Data Results ->")
print(x_val, y_val)
```

```python
#show results for training dataset
print("Test Data Results ->")
print(x_test, y_test)
```
